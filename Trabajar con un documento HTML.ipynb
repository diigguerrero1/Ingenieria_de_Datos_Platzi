```python
import requests
import bs4

response = requests.get('https://www.eluniversal.com.co/')
```

```python
print(response.text)
```

```python
soup = bs4.BeautifulSoup(response.text, 'html.parser')

print(soup.select('.vertical.menu.menu-3 a'))
```

```python
articles_links = soup.select('.vertical.menu.menu-3 a')
articles = [article['href'] for article in articles_links]

for article in articles:
    print(article)
```
